PART - 2

```python
df1['Gender'].replace('M',1,inplace=True)
df1['Gender'].replace('F',0,inplace=True)
```

```python
df1.Gender.unique()
```

```python
df1['MaritalStatus'].replace('M',1,inplace=True)
df1['MaritalStatus'].replace('S',0,inplace=True)
```

```python
df1.head()
```

```python
len(df1.CountryRegionName.unique())
```

```python
ohe_countries=pd.get_dummies(df1['CountryRegionName'],drop_first=True)
df1.drop(columns=['CountryRegionName'],axis=1,inplace=True)
df1=pd.concat([ohe_countries,df1],axis=1)
```

```python
df1.head()
```

```python
df1['BirthDate']= pd.to_datetime(df1['BirthDate'])

import datetime
CURRENT_TIME = datetime.datetime.now()

def get_age(birth_date,today=CURRENT_TIME):
    y=today-birth_date
    return y.days//365

df1['Age']=df1['BirthDate'].apply(lambda x: get_age(x))

df1.drop(['BirthDate'],axis=1,inplace=True)
```

```python
df1.head()
```

```python
len(df1.Education.unique())
```

```python
df1['Education'].value_counts()
```

```python
df1['Education']=df1['Education'].map({'Partial High School':1,'High School':2,'Partial College':3,'Bachelors':4,'Graduate Degree':5})
```

```python
df1['Occupation']=df1['Occupation'].map({'Manual':1,'Skilled Manual':2,'Clerical':3,'Management':4,'Professional':5})
```

```python
df1.head()
```

```python
df1.isnull().sum()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

df1[['YearlyIncome_scaled','Age_Scaled']] = scaler.fit_transform(df1[['YearlyIncome','Age']])
```

```python
df1.head()
```

```python
df1.drop(['YearlyIncome','Age'],axis=1,inplace=True)
```
