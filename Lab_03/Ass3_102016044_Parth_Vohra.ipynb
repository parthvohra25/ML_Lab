```python
df= pd.read_csv('/Users/Aaryan/Downloads/imports-85.data', delimiter=',')
df.columns= ["symboling", "normalized_losses", "make", "fuel_type", "aspiration",
                 "num_doors", "body_style", "drive_wheels", "engine_location", "wheel_base",
                 "length", "width", "height", "curb_weight", "engine_type", "num_cylinders",
                 "engine_size", "fuel_system", "bore", "stroke", "compression_ratio", "horsepower",
                 "peak_rpm", "city_mpg", "highway_mpg", "price"]

```

```python
filter=df =="?"

df.where(filter).count()
```

```python
###For normalized_losses
play=df[df['normalized_losses']!='?']
normalized_losses_median=play['normalized_losses'].median()
df['normalized_losses'].replace(to_replace="?",value= normalized_losses_median, inplace=True )

###For num_doors
play=df[df['make'].isin(["dodge"])]
play=play[play['aspiration']=='turbo']
aspiration_turbo_mode=play['num_doors'].mode()
df['num_doors'].replace(to_replace="?",value= "two", inplace=True )

###For bore
play=df[df['make']=='mazda']
play=play[play['fuel_type']=='gas']

play=play[play['bore']!='?']
play['bore']=play['bore'].astype(float)
bore_mean=play['bore'].mean()
df['bore'].replace(to_replace="?",value= bore_mean, inplace=True )

###For stroke
play=play[play['stroke']!='?']
play['stroke']=play['stroke'].astype(float)
bore_mean=play['stroke'].mean()
df['stroke'].replace(to_replace="?",value= bore_mean, inplace=True )


###For horsepower
df[df['horsepower'].isin(['?'])]

play=df[df['fuel_type'].isin(['gas'])]
play=play[play['aspiration'].isin(['std'])]
play=play[play['drive_wheels'].isin(['fwd'])]
play=play[play['fuel_system'].isin(['mpfi'])]

play=play[play['horsepower']!='?']
horsepower_mean=play['horsepower'].mean()
df['horsepower'].replace(to_replace="?",value= horsepower_mean, inplace=True )

###For peak_rpm
play=play[play['peak_rpm']!='?']
rpm_mean=play['peak_rpm'].mean()
df['peak_rpm'].replace(to_replace="?",value= rpm_mean, inplace=True )

###For Price
play=df[df['price']!='?']
play['price']=play['price'].astype(float)
price_mean=play['price'].mean()
df['price'].replace(to_replace="?",value= price_mean, inplace=True )

```

```python
df['num_doors'].replace(to_replace="two",value= 2, inplace=True )
df['num_doors'].replace(to_replace="four",value= 4, inplace=True )

df['num_cylinders'].replace(to_replace="four",value= 4, inplace=True )
df['num_cylinders'].replace(to_replace="six",value= 6, inplace=True )
df['num_cylinders'].replace(to_replace="five",value= 5, inplace=True )
df['num_cylinders'].replace(to_replace="three",value= 3, inplace=True )
df['num_cylinders'].replace(to_replace="twelve",value= 12, inplace=True )
df['num_cylinders'].replace(to_replace="two",value= 2, inplace=True )
df['num_cylinders'].replace(to_replace="eight",value= 8, inplace=True )

df=pd.get_dummies(df, columns=["body_style","drive_wheels"])

label_encoder = preprocessing.LabelEncoder()
df['make']= label_encoder.fit_transform(df['make'])
df['aspiration']= label_encoder.fit_transform(df['aspiration'])
df['engine_location']= label_encoder.fit_transform(df['engine_location'])
df['fuel_type']= label_encoder.fit_transform(df['fuel_type'])


df['fuel_system'].replace(to_replace=['mpfi','spfi'],value= 1, inplace=True )
df['fuel_system'].replace(to_replace=['2bbl', 'mfi', '1bbl', '4bbl', 'idi', 'spdi'],value= 0, inplace=True )

df['engine_type'].replace(to_replace=['dohc', 'ohcv', 'ohc', 'ohcf', 'dohcv'],value= 1, inplace=True )
df['engine_type'].replace(to_replace=['l', 'rotor'],value= 0, inplace=True )
```

```python
scaler=MinMaxScaler()

df_scaled=scaler.fit_transform(df)

dfs=pd.DataFrame(df_scaled)
dfs.columns=df.columns


dfx=dfs.drop("price", axis=1)
dfy=pd.DataFrame(dfs['price'])
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(dfx,dfy, test_size=0.3)

alg=LinearRegression()

alg.fit(xtrain, ytrain)

predict=alg.predict(xtest)

print('mean_squared_error \n',metrics.mean_squared_error(predict,ytest))
print('\nmean_absolute_error \n',metrics.mean_absolute_error(predict,ytest))
print('\nmean_squared_error\n', np.sqrt(metrics.mean_squared_error(predict,ytest)))
print('\nr2_score\n', metrics.r2_score(ytest, predict))
```
